# Linked List

1. Reverse Linked List — LeetCode 206

```python
def reverseList(self, head):

    prev,curr = None,head
    while curr:
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt
    return prev
```

2. Middle of the Linked List — LeetCode 876

```python
def middleNode(self, head):
    slow = fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    return slow
```

3. Merge Two Sorted Lists — LeetCode 21

```python
def mergeTwoLists(self, list1, list2):
    dummy = ListNode()
    current = dummy  

    while list1 and list2:
        if list1.val <= list2.val:
            current.next = list1
            list1 = list1.next
        else:
            current.next = list2
            list2 = list2.next
        current = current.next

    if list1:
        current.next = list1
    elif list2:
        current.next = list2

    return dummy.next
```

4. Linked List Cycle — LeetCode 141

```python
def hasCycle(self, head):

    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False
```

5. Linked List Cycle II — LeetCode 142

```python
def detectCycle(self, head):

    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            new  = head
            while new != slow:
                new = new.next
                slow = slow.next
            return new
    return None
```

6. Remove Nth Node From End — LeetCode 19

```python
def removeNthFromEnd(self, head, n):

    dummy = ListNode(0,head)
    right = head
    left = dummy

    while n>0 and right:
        right = right.next
        n -= 1
    
    while right:
        right = right.next
        left = left.next

    left.next = left.next.next
    return dummy.next
    
```

7. Palindrome Linked List — LeetCode 234

```python
 def isPalindrome(self, head):

    values = []
    current = head

    while current:
        values.append(current.val)
        current = current.next

    return values == values[::-1]
```

8. Intersection of Two Linked Lists — LeetCode 160

```python
 def getIntersectionNode(self, headA, headB):
    p1, p2 = headA, headB

    while p1 != p2:
        p1 = p1.next if p1 else headB
        p2 = p2.next if p2 else headA

    return p1
```

9. Reorder List — LeetCode 143

```python
 def reorderList(self, head):

    
    slow,fast = head,head.next
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    second = slow.next
    head1  = slow.next = None
    while second:
        tmp = second.next
        second.next = head1
        head1 = second
        second = tmp

    first,second = head, head1
    while second:
        tmp1,tmp2 = first.next,second.next
        first.next = second
        second.next = tmp1
        first,second = tmp1,tmp2
```
