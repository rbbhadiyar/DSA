1. Implement Queue using Stacks — LeetCode 232

```python
def __init__(self):
    self.s1 = []
    self.s2 = []

def push(self, x):
    self.s1.append(x)

def pop(self):
    self.peek()
    return self.s2.pop()

def peek(self):
    if not self.s2:
        while self.s1:
            self.s2.append(self.s1.pop())
    return self.s2[-1]

def empty(self):
    return not self.s1 and not self.s2
```

2. Number of Recent Calls — LeetCode 933

```python
def __init__(self):
        self.q = deque()

def ping(self, t):
    self.q.append(t)
    while self.q[0] < t - 3000:
        self.q.popleft()
    return len(self.q)
```

3. Time Needed to Buy Tickets — LeetCode 2073

```python
def timeRequiredToBuy(self, tickets, k):
    time = 0
    for i in range(len(tickets)):
        if i <= k:
            time += min(tickets[i], tickets[k])
        else:
            time += min(tickets[i], tickets[k] - 1)
    return time
```

4. First Unique Character in a String — LeetCode 387

```python
def firstUniqChar(self, s):
    count = Counter(s)
    q = deque()

    for i, ch in enumerate(s):
        q.append((ch, i))
        while q and count[q[0][0]] > 1:
            q.popleft()

    return q[0][1] if q else -1
```

5. Reveal Cards in Increasing Order — LeetCode 950

```python
def deckRevealedIncreasing(self, deck):
    deck.sort()
    q = deque()

    for card in reversed(deck):
        if q:
            q.appendleft(q.pop())
        q.appendleft(card)
    
    return list(q)
```

6. Implement Stack using Queues — LeetCode 225

```python
def __init__(self):
        self.q = deque()

def push(self, x):
    self.q.append(x)
    for _ in range(len(self.q) - 1):
        self.q.append(self.q.popleft())

def pop(self):
    return self.q.popleft()

def top(self):
    return self.q[0]

def empty(self):
    return len(self.q) == 0
```

8. Reverse a Queue

```python
def reverse_queue(q):
    stack = []

    while q:
        stack.append(q.popleft())

    while stack:
        q.append(stack.pop())

    return q
```

9. Check if a String is Palindrome using Queue

```python
def is_palindrome(s):
    q = deque()
    for ch in s:
        q.append(ch)

    while len(q) > 1:
        if q.popleft() != q.pop():
            return False

    return True
```

10. Generate Binary Numbers from 1 to N using Queue

```python

def generate_binary(n):
    q = deque()
    q.append("1")

    for _ in range(n):
        curr = q.popleft()
        print(curr)
        q.append(curr + "0")
        q.append(curr + "1")
```
