1. Valid Parentheses – LeetCode 20

```python
def isValid(self, s):
        stack = []
        for char in s:
            if char == '(':
                stack.append(')') 
            elif char == '{':
                stack.append('}') 
            elif char == '[':
                stack.append(']') 
            elif not stack or stack.pop() !=char:
                return False
        return not stack
    
```

2. Min Stack – LeetCode 155

```python


def __init__(self):
    self.stack = []
    self.minStack = []

def push(self, val):
    self.stack.append(val)

    if not self.minStack or val <= self.minStack[-1]:
        self.minStack.append(val)

def pop(self):
    val = self.stack.pop()
    if val == self.minStack[-1]:
        self.minStack.pop()

def top(self):
    return self.stack[-1]

def getMin(self):
    return self.minStack[-1]
```

3. Evaluate Reverse Polish Notation – LeetCode 150

```python

def evalRPN(self, tokens):
    
    stack = []

    for t in tokens:
        if t in "+-*/":
            b = stack.pop()
            a = stack.pop()

            if t == '+':
                stack.append(a + b)
            elif t == '-':
                stack.append(a - b)
            elif t == '*':
                stack.append(a * b)
            else:   
                stack.append(int(float(a) / b)) 

        else:
            stack.append(int(t))

    return stack[0]
```

4. Next Greater Element I – LeetCode 496

```python

def nextGreaterElement(self, nums1, nums2):
    
    stack = []
    nge = {} 

    for n in nums2:
        while stack and stack[-1] < n:
            nge[stack.pop()] = n
        stack.append(n)

    return [nge.get(n, -1) for n in nums1]
```

5. Next Greater Element II – LeetCode 503

```python

def nextGreaterElements(self, nums):
    
    n = len(nums)
    res = [-1] * n
    stack = []

    for i in range(2 * n):
        num = nums[i % n]
        while stack and nums[stack[-1]] < num:
            res[stack.pop()] = num
        if i < n:
            stack.append(i)

    return res
```

6. Daily Temperatures – LeetCode 739

```python

def dailyTemperatures(self, temps):
    res = [0] * len(temps)
    stack = [] 
    for i, t in enumerate(temps):
        while stack and temps[stack[-1]] < t:
            idx = stack.pop()
            res[idx] = i - idx
        stack.append(i)

    return res
```

8. Implement Queue using Stacks – LeetCode 232

```python
def __init__(self):
    self.s1 = []
    self.s2 = []

def push(self, x):
    self.s1.append(x)

def pop(self):
    self.peek()
    return self.s2.pop()

def peek(self):
    if not self.s2:
        while self.s1:
            self.s2.append(self.s1.pop())
    return self.s2[-1]

def empty(self):
    return not self.s1 and not self.s2
```

9. Remove All Adjacent Duplicates in String – LeetCode 1047

```python
def removeDuplicates(self, s):
    stack = []
    for char in s:
        if stack and stack[-1]==char:
            stack.pop()
        else:
            stack.append(char)
            
    return "".join(stack)
```

10. Largest Rectangle in Histogram – LeetCode 84

```python
def largestRectangleArea(self, heights):
    stack = []
    maxArea = 0

    for i, h in enumerate(heights + [0]):
        while stack and heights[stack[-1]] > h:
            height = heights[stack.pop()]
            width = i if not stack else i - stack[-1] - 1
            maxArea = max(maxArea, height * width)
        stack.append(i)

    return maxArea
```
